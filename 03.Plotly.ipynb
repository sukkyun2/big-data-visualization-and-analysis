The **Superstore** Sales dataset contains sales transactions from a retail store, including information such as product categories, sales amounts, and customer details.

Reference: https://www.kaggle.com/datasets/rohitsahoo/sales-forecasting?select=train.csv

```python
import kagglehub

# 데이터셋 다운로드
path = kagglehub.dataset_download("rohitsahoo/sales-forecasting")
path
```

```python
import pandas as pd
import os

df = pd.read_csv(os.path.join(path, "train.csv"))
df.head()
```

```python
# 데이터셋 상세정보 확인
df.info()
```

## Apply basic charts

### Sales Trend by Category Over Time

Using Line Chart

#### **X-Axis: Order Date**
The date data, in the format mm/dd/yy, is grouped by month and displayed accordingly.

#### **Y : Sales**
The y-axis represents the sales amount in dollars This is the amount of money earned from sales on each given date.

#### Result
The highest sales occurred in November 2018

```python
import plotly.express as px
import pandas as pd

# 날짜 포맷팅
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%d/%m/%Y')

# 월별 판매량 계산
df_monthly = df.resample('ME', on='Order Date').agg({'Sales': 'sum'}).reset_index()

fig = px.line(df_monthly,
              x='Order Date',
              y='Sales',
              title='월별 판매량',
              labels={'Sales': '판매량 ($)', 'Order Date': '월'})

fig.show()
```

## Apply statistical charts

### Sales by City

using histogram

#### **X-axis: City**
This axis represents different cities, categorizing the data based on the city names. Each city is plotted along the x-axis, and each city will have a corresponding sales value (sum of sales) on the y-axis.

#### **Y-axis: "Sales ($)**
The y-axis represents the total sales (in dollars) for each city. It shows the aggregated sales amount for each city, calculated using the histfunc="sum" parameter. This means that the chart will sum up all the sales in each city and display the total for that city.

#### **Result**
New York City has the highest sales.

```python
import plotly.express as px

# 히스토그램 생성
fig = px.histogram(df,
                   x="City",
                   y="Sales",
                   histfunc="sum",
                   title="도시별 판매량",
                   labels={'Sales': '판매량 ($)', 'City': '도시'})

fig.show()
```
